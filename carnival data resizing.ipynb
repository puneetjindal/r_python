```python
import pandas as pd
import openpyxl as xl
```

```python
data = pd.read_excel('Carnival_RawData.xlsx')
data = pd.DataFrame(data)
tag = pd.read_excel('Carnival_OfferTags.xlsx')
tag = pd.DataFrame(tag)
```

```python
group_list = pd.unique(tag['Group'])[1:]
for y in group_list:
    data[str(y)] = False
```

```python
def contain_join(row):
    for x in str(row['Available Offer']).split('|'):
        if not tag[tag['Unique Offers']==x].empty:
            row[unicode(tag[tag['Unique Offers']==x].iloc[0]['Group'])] = True
    return row
```

```python
df_export = data.apply(func=contain_join,axis=1)
```
